```python
import pandas as pd
from bs4 import BeautifulSoup as bs
from csv import writer
```

```python
# stop words list
stop_words = "sw.xml"
with open(stop_words, "r") as f:
    content = f.read()
    sw_content = bs(content, "lxml")
```

```python
# this is a path of the csv that we are going to store all the court cases.
make_csv = "data.csv"
```

```python
import os
# the path is for the domstolar corpus, you can fetch the data from http://malfong.is The Icelandic Gigaword corpus, 
# put the directory in the same place as this script
for root, dirs, files in os.walk('domstolar/'):
    for file in files:
        tmp_list = []
        with open(os.path.join(root, file), "r") as auto:
            content = auto.read()
            bs_content = bs(content, "lxml")
        all_word = bs_content.find_all('w')
        
        list_of_words = [ words.get('lemma').lower() \
                 for words in all_word if not any(char.isdigit() \
                 for char in words.get('lemma')) and len(words.get('lemma'))> 2 \
                 and not any(char for char in words.get('lemma') if char == ".")]
        
        all_stop_words = sw_content.find_all('td')
        list_of_stop_words = [ words.text.lower() for words in all_stop_words if words.text.isalpha()]
        stop_words_removed_list = [ words for words in list_of_words if words not in list_of_stop_words ]
        final_sentence = ' '.join(stop_words_removed_list)
        tmp_list.append(os.path.join(root, file))
        tmp_list.append(final_sentence)
        
        with open(make_csv, 'a+', newline='') as write_obj:
            csv_writer = writer(write_obj)
            csv_writer.writerow(tmp_list)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv(make_csv, sep=",", names = ['file_name','content'])
```

```python
true_k = 10
```

```python
model1 = KMeans(n_clusters=true_k, init='k-means++', max_iter=100, n_init=1)
```

```python
model1.fit(X1)
```

```python
order_centroids = model1.cluster_centers_.argsort()[:, ::-1]
```

```python
terms = vectorizer.get_feature_names()
```

```python
for i in range(true_k):
    print("Cluster %d:" % i),
    for ind in order_centroids[i, :10]:
        print(" %s " % terms[ind])
```

```python
import pandas as pd;
import numpy as np;
import scipy as sp;
import sklearn;
import sys;
from nltk.corpus import stopwords;
import nltk;
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer;
from sklearn.decomposition import NMF;
from sklearn.preprocessing import normalize;
import pickle;
```

```python
vectorizer = CountVectorizer(analyzer='word', max_features=5000);
x_counts = vectorizer.fit_transform(data['content']);
```

```python
transformer = TfidfTransformer(smooth_idf=False);
x_tfidf = transformer.fit_transform(x_counts);
```

```python
xtfidf_norm = normalize(x_tfidf, norm='l1', axis=1)
```

```python
num_topics = 10;
```

```python
model2 = NMF(n_components=num_topics, init='nndsvd');
```

```python
model2.fit(xtfidf_norm)
```

```python
def get_nmf_topics(model, n_top_words):
    
    #the word ids obtained need to be reverse-mapped to the words so we can print the topic names.
    feat_names = vectorizer.get_feature_names()
    
    word_dict = {};
    for i in range(num_topics):
        
        #for each topic, obtain the largest values, and add the words they map to into the dictionary.
        words_ids = model.components_[i].argsort()[:-20 - 1:-1]
        words = [feat_names[key] for key in words_ids]
        word_dict['Topic # ' + '{:02d}'.format(i+1)] = words;
    
    return pd.DataFrame(word_dict);
```

```python
get_nmf_topics(model2, 20)
```
